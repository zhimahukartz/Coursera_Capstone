# get major U.S and Canadian cities' according to their population

## we get the information from Wikipedia, and we need to use beautifulsoup to scrape data

### load the website

```python
url = 'https://en.wikipedia.org/wiki/List_of_North_American_cities_by_population'
html = urlopen(url)
soup = BeautifulSoup(html, 'lxml')
table = soup.find('table', class_='wikitable sortable')
```

### scrape the table data into DataFrame

```python
l = []
for tr in table.find_all('tr'):
    td = tr.find_all('td')
    row = [tr.text for tr in td]
    l.append(row)
df = pd.DataFrame(l,columns=['Rank','City','Image', 'Country', 'Population','Year'])
df
```

#### drop column image, which will not be used

```python
df1 = df.drop('Image', axis=1)
```

#### drop the first row

```python
df2 = df1.drop(0,axis=0)
df2.head()
```

#### remove all \n in the DataFrame

```python
df3 = df2.apply(lambda x: x.str.strip('\n'))
```

#### use regular expression to remove all [] and content inside of it

```python
df3['Population'] = df3['Population'].apply(lambda x: re.sub(r'\[.*?\]', '', x))
df3['Year'] = df3['Year'].apply(lambda x: re.sub(r'\[.*?\]', '', x))
```

```python
df3.head()
```

#### we are only interested in U.S cities, so keep them and remove other countries

```python
df4 = df3[df3['Country'].str.contains('United States')]
```

```python
df5 = df4
```

#### reset the ranking

#### combine column city and country and give a new column location, which will be more precise to get grogrophic location

```python
df5['Location'] = df5['City'] + ','+ df5['Country']
```

#### use Nominatim to get geolocation data of the cities and create latitude and longitude columns in dataframe

```python
list1 =[]
list2 =[]
list3=[]
for location in df5.Location:
    address = location

    geolocator = Nominatim()
    location = geolocator.geocode(address)
    list3.append(location)
    latitude = location.latitude
    list1.append(latitude)
    longitude = location.longitude
    list2.append(longitude)
    print('The geograpical coordinate of {} are {}, {}.'.format(location, latitude, longitude))
df5['Latitude'] = list1
df5['Longitude'] = list2
df5['Location'] = list3
```

### Reset index

## Use Foursquare API to get how many Chinese restaurants

#### load Foursqaure API information

```python
CLIENT_ID = '##' # your Foursquare ID
CLIENT_SECRET = '##' # your Foursquare Secret
VERSION = '20180605' # Foursquare API version

print('Your credentails:')
print('CLIENT_ID: ' + CLIENT_ID)
print('CLIENT_SECRET:' + CLIENT_SECRET)
```

#### create a new column in city_data called 'how_many' to indicate how many Chinese restaurants are within 3000m of the cities geolocation. Use Foursquare's search function to search Chinese restaurant. 

```python
df12['how_many'] = 'None'
list=[]
for index, row in df12.iterrows():
    neighborhood_latitude = row['Latitude']
    neighborhood_longitude = row['Longitude']
    categoryId = '4bf58dd8d48988d145941735'
    LIMIT =50
    radius = 2000
    url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}&categoryId={}'.format(
    CLIENT_ID,
    CLIENT_SECRET,
    VERSION,
    neighborhood_latitude,
    neighborhood_longitude,
    radius,
    LIMIT,
    categoryId)
    results = requests.get(url).json()
    names = results['response']['venues']
    names = json_normalize(names)
    filtered_columns = ['name', 'location.lat', 'location.lng']
    nearby_venues = names.loc[:, filtered_columns]

    how_many = len(nearby_venues.name)
    list.append(how_many)
df12['how_many'] = list
```

```python
df12
```

```python
df13 = df12
```

#### use population_3km / how_many to get how many people are haivng one Chinese Restaurants

```python
df13['person/restaurant'] = df13['population_3km']/df13['how_many']
```

```python
df13
```

#### sort values

```python
df13.sort_values('person/restaurant')
```

#### Mapping person/restaurant to the U.S map

```python
us_map = folium.Map(location=[39.5,-98.35],zoom_start=4)
for lat, lng, loca, density in zip(df13['Latitude'], df13['Longitude'], df13['New_Location'], df13['person/restaurant']):
    label = '{}'.format(loca)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=density/400,
        popup=label,
        color='red',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(us_map)  
us_map
```
